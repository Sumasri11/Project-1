```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("/Users/sumasrijasti/Downloads/customer_booking.csv", encoding='latin1')
```

```python
data.head()
```

```python
# Visualize distribution of number of passengers
plt.figure(figsize=(10, 6))
sns.histplot(data['num_passengers'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Number of Passengers')
plt.xlabel('Number of Passengers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize distribution of sales channels
plt.figure(figsize=(10, 6))
ax = sns.countplot(x='sales_channel', data=data)
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 5), 
                 textcoords = 'offset points')

plt.title('Distribution of Sales Channels')
plt.xlabel('Sales Channel')
plt.show()
```

```python
# Visualize count of trip types
plt.figure(figsize=(8, 6))
ax = sns.countplot(x='trip_type', data=data)
plt.title('Count of Trip Types')
plt.xlabel('Trip Type')
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha = 'center', va = 'center', 
                 xytext = (0, 5), 
                 textcoords = 'offset points')

plt.show()
```

```python
# Visualize distribution of purchase lead time
plt.figure(figsize=(10, 6))
sns.histplot(data['purchase_lead'], bins=30, kde=True, color='green')
plt.title('Distribution of Purchase Lead Time')
plt.xlabel('Purchase Lead Time (days)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize distribution of flight durations
plt.figure(figsize=(10, 6))
sns.histplot(data['flight_duration'], bins=30, kde=True, color='orange')
plt.title('Distribution of Flight Durations')
plt.xlabel('Flight Duration (hours)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Aggregate count of completed bookings
booking_complete_count = data['booking_complete'].value_counts().sort_index()

# Visualize count of completed bookings as a line graph
plt.figure(figsize=(10, 6))
booking_complete_count.plot(kind='line', marker='o', color='blue')
plt.title('Count of Completed Bookings')
plt.xlabel('Booking Completed')
plt.ylabel('Count')
plt.xticks([0, 1], ['Incomplete', 'Complete'])
plt.grid(True)
plt.show()
```
